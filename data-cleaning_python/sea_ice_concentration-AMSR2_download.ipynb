### Sea Ice Concentration daily average

Data available from AMSR2: https://nsidc.org/data/au_si12/versions/1 <br>
The nominal spatial resolution of the polar grids is 12.5 km. However, because the polar grids are
not equal area, the actual resolution varies by latitude. <br>


User Guide found here: https://nsidc.org/sites/default/files/au_si12-v001-userguide_1.pdf

NSIDC Sea Ice Polar Stereographic North <br>
EPSG code: 3411

Sea ice concentration values (units: percent):
- 0: Open Water
- 1 - 100: Percent sea ice concentration
- 110: Missing data
- 120: Land
